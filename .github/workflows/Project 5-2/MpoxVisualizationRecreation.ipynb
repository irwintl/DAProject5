Mpox visualization recreation:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("owid-mpox-data.csv")
```

```python
df.head(2)
```

Anticipated Process
1. Set location == "World" / "North America"
2. "new_cases_smoothed" is the seven-day average, so we will do a line graph of that variable

```python
world_cases = df[df["location"] == "World"]
na_cases = df[df["location"] == "North America"]
plt.plot(world_cases['date'], world_cases['new_cases_smoothed'], label = "World")
plt.plot(na_cases['date'], na_cases['new_cases_smoothed'], label = "North America")

plt.style.use('tableau-colorblind10')

plt.xlabel('Date')
plt.ylabel('Cases')
plt.title('Mpox 7-day rolling average')



plt.legend()

plt.show()
```

```python
df_sorted = df[['location', 'new_cases_smoothed', "date" ]]
df_sorted = df_sorted[(df_sorted["location"] == "World") | (df_sorted["location"] == "North America")]
df_sorted = df[['col1', 'col2']]
```
